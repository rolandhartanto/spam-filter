## Import

```python
import os, stat, string, re
import pandas as pd
import numpy as np
import nltk
import math

from collections import Counter
from nltk import word_tokenize, WordNetLemmatizer
from nltk.corpus import wordnet, stopwords
```

### Load Data

```python
def load_data(src):
    emails = ['' for _ in range(2500)]
    if not os.path.exists(src):
        os.makedirs(src)
    files = os.listdir(src)
    for file in files:
        idx = file.replace('TRAIN_','').replace('.eml','')
        path = os.path.join(src, file)
        info = os.stat(path)
        if not stat.S_ISDIR(info.st_mode):
            fp = open(path, encoding='utf-8', errors='ignore')
            body = fp.read()
            fp.close()
            
            emails[int(idx) - 1] = body
    return emails
    

    
labels = pd.read_csv('spam-mail.tr.label')
labels = labels['Prediction'].tolist()
emails = load_data('TR-mails-extracted/TR')
```

## Preprocessing

### Remove Punctuations

```python
trans = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
emails = [email.translate(trans) for email in emails]
```

### Tokenize

```python
tokens = [word_tokenize(email) for email in emails]
```

### Lemmatization

```python
tagged = [nltk.pos_tag(token) for token in tokens]
```

```python
def get_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

lemmatizer = WordNetLemmatizer()
tokens = [[lemmatizer.lemmatize(word[0].lower(), pos=get_pos(word[1])) for word in token] for token in tagged]
```

## Feature Selection

### Stop Words Elimination

```python
stop_words = set(stopwords.words('english'))
tokens = [[word for word in token if not word in stop_words] for token in tokens]
tokens[1]
```
